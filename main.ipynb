```python
from sklearn.datasets import make_blobs
```

```python
X,Y=make_blobs(n_samples=400,n_features=2,centers=2)
```

```python
test_fruit=np.array([8,10])
plt.scatter(test_fruit[0],test_fruit[1],color='green',marker='*')

plt.scatter(X[:,0],X[:,1],c=Y)
```

```python
def dist(a1,a2):
    return np.sum((a1-a2)**2)**.5
def KNN(X,Y,test_point,k=1):
    m=X.shape[0]
    vals=[]
    for i in range(m):
        d=dist(test_point,X[i])
        vals.append((d,Y[i]))
    vals=sorted(vals)
    vals=np.array(vals[:k])
    b=np.unique(vals[:,1],return_counts=True)
    idx=np.argmax(b[1])
    pred=b[0][idx]
    return int(pred)
```

```python
KNN(X,Y,test_fruit)
```

```python
p=Path("./trial/")
dirs = p.glob("*")

image_data=[]
labels=[]

labels_dict={"Pikachu":0,"Bulbasaur":1,"Charmander":2}

image_paths = []


for folder_dir in dirs:
    label = str(folder_dir).split("\\")[-1]
    
    cnt=0
    
    print(folder_dir.name)
    
    for img_path in folder_dir.glob('*.jpg'):
        
        #img=image.load(img_path,target_size=(100,100))
        imag=cv2.imread(str(img_path))
        img1=cv2.resize(imag,(100,100))
        img=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
        img_array=np.array(img)
        image_data.append(img_array)
        labels.append(labels_dict[label])
        cnt+=1
    print(cnt)
    
        
```
